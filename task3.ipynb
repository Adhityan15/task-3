```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("C:\\Users\\adhit\\Downloads\\householdtask3.csv")
#csv file readed
```

```python
#visualising data using pandas data frame
pd.DataFrame(df)
```

```python
#bar chart
plt.figure(figsize=(10,7))
plt.bar(df["year"],df["expenditure"],color="blue",label="household")
plt.xlabel("year")
plt.ylabel("expence")
plt.title("houshold expences")
plt.legend()
plt.show
           
           
```

```python
plt.figure(figsize=(10,5))
plt.plot(df["year"],df["income"],marker="o",color="blue",label="household")
plt.ylabel("year")
plt.ylabel("income")
plt.title("income")
plt.legend()
plt.show
       
```
